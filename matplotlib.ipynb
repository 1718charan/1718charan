```python
import matplotlib.pyplot as plt
import numpy as np
profit=[30,30,25,15.65]
labels=["bmw","rollsroyce","ferrari","lamborghini"]
explode=[0.1,0.02,0,0.05]
plt.pie(profit,labels=labels,counterclock=False,explode=explode,autopct="%.2f%%")
plt.show()
```
